```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from tqdm import tqdm
```

```python
### maximum number in python float64 ###
np.finfo(np.float64).max
```

```python
# example: what happens if we enter a number larger than the maximum?
a = 1.8e+308
print(a)
```

```python
# smalles subnormal number in python float 64
np.finfo(np.float64).smallest_subnormal
```

```python
# Example: what happens if we enter a number that is smaller than the smallest?
b = 6e-324
print(b)
```

```python
# machine epsilon
eps = np.finfo(np.float64).eps

print(eps)
```

```python
# example 1
c = 1 + eps - 1
print(c)
```

```python
# example 2
c = 1 + 0.5*eps - 1
print(c)
```

```python
# example 3
c = 10 + eps - 10
print(c)
```

```python
# example 4
d = 0.3 + 0.1 - 0.3 - 0.1 
print(d)
```
